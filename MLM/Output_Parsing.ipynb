```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("test_unlike_probabilities.csv")

# Initialize a list to store the labels
labels = []

# Iterate through the rows of the DataFrame
for index, row in df.iterrows():
    generated_text = str(row['Generated Text']).lower()  # Convert to lowercase for case-insensitive matching
    if 'unlike' in generated_text:
        labels.append(1)  # Append 1 if 'unlike' is found
    elif 'similar' in generated_text:
        labels.append(0)  # Append 0 if 'similar' is found
    else:
        labels.append(None)  # Append None if neither 'unlike' nor 'similar' is found

# Print the labels list
print(labels)
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import precision_recall_fscore_support
```

```python
df_label = pd.read_csv("test.csv")
```

```python
accuracy = accuracy_score(df_label["Label"], labels)
print("Accuracy:", accuracy)
```

```python
report = classification_report(df_label["Label"], labels, target_names=['True', 'False'])
```

```python
precision, recall, f1_score, support = precision_recall_fscore_support(df_label["Label"], labels)
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("test.csv")

# Initialize a list to store the labels
labels = []

# Iterate through the rows of the DataFrame
for index, row in df.iterrows():
    generated_text = str(row['Body']).lower()  # Convert to lowercase for case-insensitive matching
    if 'unlike' in generated_text:
        labels.append(1)  # Append 1 if 'unlike' is found
    elif 'similar' in generated_text:
        labels.append(0)  # Append 0 if 'similar' is found
    else:
        labels.append(2)  # Append None if neither 'unlike' nor 'similar' is found

# Print the labels list
print(labels)
```

```python


# Count occurrences of each value
count_0 = labels.count(0)
count_1 = labels.count(1)
count_2 = labels.count(2)

print("Count of 0:", count_0)
print("Count of 1:", count_1)
print("Count of 2:", count_2)
```
